```python
# Road Damage Detection using Vision Transformers (ViT)
# This code implements a multi-class classification system for detecting different types of road damages

import os
import glob
import zipfile
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import cv2
from sklearn.metrics import confusion_matrix, classification_report

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
from torchvision.models import vit_b_16, ViT_B_16_Weights
import torchvision.transforms.functional as TF

from tqdm import tqdm
import pandas as pd
from google.colab import drive

# Set random seeds for reproducibility
torch.manual_seed(42)
np.random.seed(42)

# Mount Google Drive for Colab if needed
def mount_drive():
    """Mount Google Drive to access datasets stored there"""
    print("Mounting Google Drive...")
    drive.mount('/content/drive')
    print("Google Drive mounted successfully!")

# Data extraction functions
def extract_dataset(zip_path, extract_to):
    """Extract the dataset from zip file to the specified directory"""
    print(f"Extracting dataset from {zip_path} to {extract_to}")

    # Create extraction directory if it doesn't exist
    os.makedirs(extract_to, exist_ok=True)

    # Extract the ZIP file
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)

    print("Dataset extracted successfully!")

    # List extracted directories to help verify the structure
    print("\nExtracted directory structure:")
    for root, dirs, files in os.walk(extract_to, maxdepth=3):
        level = root.replace(extract_to, '').count(os.sep)
        indent = ' ' * 4 * level
        print(f"{indent}{os.path.basename(root)}/")
        if level < 2:  # Only show subdirectories and a count of files at lower levels
            for d in dirs:
                print(f"{indent}    {d}/")
            if files:
                print(f"{indent}    ({len(files)} files)")

# Custom Dataset class for the road damage images
class RoadDamageDataset(Dataset):
    def __init__(self, root_dir, transform=None, subset='train'):
        """
        Args:
            root_dir (string): Directory with all the images.
            transform (callable, optional): Optional transform to be applied on a sample.
            subset (string): 'train', 'test', or 'val' subset
        """
        self.root_dir = root_dir
        self.transform = transform
        self.subset = subset

        # Define the classes
        self.classes = ['crack', 'normal', 'pothole']
        self.class_to_idx = {cls_name: i for i, cls_name in enumerate(self.classes)}

        # Get all image paths and labels
        self.samples = []

        # Check if root directory exists
        if not os.path.exists(root_dir):
            print(f"Error: Root directory {root_dir} does not exist!")
            print("Please check your dataset path and structure.")
            return

        # First, try the expected directory structure: root_dir/class/subset/images
        expected_structure_found = False
        for class_name in self.classes:
            class_path = os.path.join(root_dir, class_name, subset)
            if os.path.exists(class_path):
                expected_structure_found = True
                image_count = 0
                for img_path in glob.glob(os.path.join(class_path, "*.jpg")) + \
                               glob.glob(os.path.join(class_path, "*.jpeg")) + \
                               glob.glob(os.path.join(class_path, "*.png")):
                    self.samples.append((img_path, self.class_to_idx[class_name]))
                    image_count += 1
                print(f"Found {image_count} images in {class_path}")

        # If the expected structure is not found, try alternative structures
        if not expected_structure_found:
            print("Expected directory structure not found. Trying alternative structures...")

            # Try structure: root_dir/subset/class/images
            for class_name in self.classes:
                class_path = os.path.join(root_dir, subset, class_name)
                if os.path.exists(class_path):
                    print(f"Found alternative structure: {class_path}")
                    image_count = 0
                    for img_path in glob.glob(os.path.join(class_path, "*.jpg")) + \
                                   glob.glob(os.path.join(class_path, "*.jpeg")) + \
                                   glob.glob(os.path.join(class_path, "*.png")):
                        self.samples.append((img_path, self.class_to_idx[class_name]))
                        image_count += 1
                    print(f"Found {image_count} images in {class_path}")

            # If still no images found, try looking directly for subdirectories that match class names
            if not self.samples:
                print("Searching for class directories anywhere in the dataset...")
                for root, dirs, _ in os.walk(root_dir):
                    for class_name in self.classes:
                        if class_name in dirs:
                            class_path = os.path.join(root, class_name)
                            print(f"Found class directory: {class_path}")
                            image_count = 0
                            for img_path in glob.glob(os.path.join(class_path, "*.jpg")) + \
                                           glob.glob(os.path.join(class_path, "*.jpeg")) + \
                                           glob.glob(os.path.join(class_path, "*.png")):
                                self.samples.append((img_path, self.class_to_idx[class_name]))
                                image_count += 1
                            print(f"Found {image_count} images in {class_path}")

        # Final attempt: just try to find images directly within the root directory
        if not self.samples:
            print(f"No structured dataset found. Looking for images directly in {root_dir}...")
            for img_path in glob.glob(os.path.join(root_dir, "**/*.jpg"), recursive=True) + \
                           glob.glob(os.path.join(root_dir, "**/*.jpeg"), recursive=True) + \
                           glob.glob(os.path.join(root_dir, "**/*.png"), recursive=True):
                # Try to determine class from path
                for class_name in self.classes:
                    if class_name in img_path.lower():
                        self.samples.append((img_path, self.class_to_idx[class_name]))
                        break
            print(f"Found {len(self.samples)} images with class information in filename")

        if len(self.samples) == 0:
            print(f"WARNING: No images found for {subset} subset! Please check your dataset path and structure.")
            print(f"Current root directory: {root_dir}")
            print("Available files and directories:")
            try:
                for item in os.listdir(root_dir):
                    item_path = os.path.join(root_dir, item)
                    if os.path.isdir(item_path):
                        print(f"  Directory: {item} (contains {len(os.listdir(item_path))} items)")
                    else:
                        print(f"  File: {item}")
            except Exception as e:
                print(f"Error listing directory contents: {e}")

            # Add a dummy sample to prevent errors
            print("Adding a dummy sample to prevent DataLoader errors")
            dummy_image = np.zeros((224, 224, 3), dtype=np.uint8)
            dummy_path = os.path.join(root_dir, "dummy_image.jpg")
            cv2.imwrite(dummy_path, dummy_image)
            self.samples.append((dummy_path, 0))  # Use class 0 (crack) for the dummy
            print("WARNING: Using dummy image for training. Results will not be meaningful!")
        else:
            print(f"Found {len(self.samples)} images in the {subset} dataset")

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img_path, label = self.samples[idx]

        try:
            # Load image
            image = Image.open(img_path).convert('RGB')

            # Apply transformations if any
            if self.transform:
                image = self.transform(image)

            return image, label
        except Exception as e:
            print(f"Error loading image {img_path}: {e}")
            # Return a black image in case of error
            dummy_image = torch.zeros(3, 224, 224) if self.transform else Image.new('RGB', (224, 224), color='black')
            return dummy_image, label

# Image preprocessing functions
def get_transforms(phase):
    """
    Define transformations for images based on phase (train/validation/test)
    """
    # Mean and std values for ImageNet normalization
    # (ViT models are typically pre-trained on ImageNet)
    mean = [0.485, 0.456, 0.406]
    std = [0.229, 0.224, 0.225]

    if phase == 'train':
        return transforms.Compose([
            transforms.Resize((224, 224)),  # ViT typically uses 224x224
            transforms.RandomHorizontalFlip(),
            transforms.RandomRotation(10),
            transforms.ColorJitter(brightness=0.2, contrast=0.2),
            transforms.ToTensor(),
            transforms.Normalize(mean=mean, std=std)
        ])
    else:  # 'val' or 'test'
        return transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=mean, std=std)
        ])

# Model definition
class RoadDamageViT(nn.Module):
    def __init__(self, num_classes=3):
        """
        Initialize the ViT model for road damage classification

        Args:
            num_classes (int): Number of classes for classification
        """
        super(RoadDamageViT, self).__init__()

        # Load pre-trained ViT model
        self.model = vit_b_16(weights=ViT_B_16_Weights.IMAGENET1K_V1)

        # Replace the classification head
        in_features = self.model.heads.head.in_features
        self.model.heads.head = nn.Linear(in_features, num_classes)

    def forward(self, x):
        return self.model(x)

# Training function
def train_model(model, dataloaders, criterion, optimizer, num_epochs=4, device='cuda'):
    """
    Train the model

    Args:
        model: The neural network model
        dataloaders: Dictionary with train and val dataloaders
        criterion: Loss function
        optimizer: Optimization algorithm
        num_epochs: Number of training epochs
        device: Device to train on ('cuda' or 'cpu')

    Returns:
        Trained model and history of metrics
    """
    model = model.to(device)

    # For tracking metrics
    history = {
        'train_loss': [],
        'train_acc': [],
        'val_loss': [],
        'val_acc': []
    }

    # For early stopping
    best_val_acc = 0.0
    patience = 5
    patience_counter = 0
    best_model_wts = model.state_dict().copy()

    for epoch in range(num_epochs):
        print(f'Epoch {epoch+1}/{num_epochs}')
        print('-' * 10)

        # Each epoch has a training and validation phase
        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()  # Set model to training mode
            else:
                model.eval()   # Set model to evaluate mode

            running_loss = 0.0
            running_corrects = 0

            # Iterate over data
            for inputs, labels in tqdm(dataloaders[phase]):
                inputs = inputs.to(device)
                labels = labels.to(device)

                # Zero the parameter gradients
                optimizer.zero_grad()

                # Forward pass (track history only in train)
                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)

                    # Backward + optimize only in training phase
                    if phase == 'train':
                        loss.backward()
                        optimizer.step()

                # Statistics
                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)
            # Check if dataset is empty
            dataset_size = len(dataloaders[phase].dataset) #Fixed: Changed 'dataloade' to 'dataloaders'
            if dataset_size == 0:
                print(f"WARNING: {phase} dataset is empty!")

                epoch_loss = 0 #Fixed: Corrected indentation
                epoch_acc = 0  #Fixed: Corrected indentation
            else:
                epoch_loss = running_loss / dataset_size
                epoch_acc = running_corrects.double() / dataset_size

            # Save metrics
            if phase == 'train':
                history['train_loss'].append(epoch_loss)
                history['train_acc'].append(epoch_acc.item() if isinstance(epoch_acc, torch.Tensor) else epoch_acc)
            else:
                history['val_loss'].append(epoch_loss)
                history['val_acc'].append(epoch_acc.item() if isinstance(epoch_acc, torch.Tensor) else epoch_acc)

            print(f'{phase} Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

            # Early stopping checkch
            if phase == 'val' and (isinstance(epoch_acc, torch.Tensor) and epoch_acc > best_val_acc or #Fixed: Changed 'tor' to 'torch'
                                  not isinstance(epoch_acc, torch.Tensor) and epoch_acc > best_val_acc):
                best_val_acc = epoch_acc
                best_model_wts = model.state_dict().copy()
                patience_counter = 0
            elif phase == 'val':
                patience_counter += 1

        print()

        # Check if we should stop training
        if patience_counter >= patience:
            print(f'Early stopping triggered after {epoch+1} epochs')
            break

    # Load best model weights
    model.load_state_dict(best_model_wts)
    return model, history

# Function to visualize training history
def plot_training_history(history):
    """
    Plot the training and validation loss and accuracy

    Args:
        history: Dictionary containing training metrics
    """
    plt.figure(figsize=(12, 5))

    # Plot training & validation loss
    plt.subplot(1, 2, 1)
    plt.plot(history['train_loss'], label='Train')
    plt.plot(history['val_loss'], label='Validation')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    # Plot training & validation accuracy
    plt.subplot(1, 2, 2)
    plt.plot(history['train_acc'], label='Train')
    plt.plot(history['val_acc'], label='Validation')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Function to evaluate model on test set
def evaluate_model(model, test_loader, criterion, device='cuda', classes=None):
    """
    Evaluate the model on the test set

    Args:
        model: Trained model
        test_loader: DataLoader for test set
        criterion: Loss function
        device: Device to evaluate on
        classes: List of class names

    Returns:
        Test accuracy and predictions
    """
    model.eval()
    test_loss = 0.0
    test_corrects = 0

    all_preds = []
    all_labels = []

    # Check if test set is empty
    if len(test_loader.dataset) == 0:
        print("WARNING: Test dataset is empty. Cannot evaluate model.")
        return 0.0, []

    # No gradient needed for evaluation
    with torch.no_grad():
        for inputs, labels in tqdm(test_loader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            loss = criterion(outputs, labels)

            test_loss += loss.item() * inputs.size(0)
            test_corrects += torch.sum(preds == labels.data)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    test_loss = test_loss / len(test_loader.dataset)
    test_acc = test_corrects.double() / len(test_loader.dataset)

    print(f'Test Loss: {test_loss:.4f} Acc: {test_acc:.4f}')

    # Compute confusion matrix
    cm = confusion_matrix(all_labels, all_preds)

    # Plot confusion matrix
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=classes, yticklabels=classes)
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

    # Print classification report
    print("\nClassification Report:")
    print(classification_report(all_labels, all_preds, target_names=classes))

    return test_acc.item(), all_preds

# Function to predict road condition (damage/non-damage) from an image
def predict_image(model, image_path, device='cuda'):
    """
    Predict the class of a single image

    Args:
        model: Trained model
        image_path: Path to the image
        device: Device to predict on

    Returns:
        Class prediction and confidence
    """
    # Classes
    classes = ['Crack', 'Normal (No Damage)', 'Pothole']

    # Check if image exists
    if not os.path.exists(image_path):
        print(f"Error: Image not found at {image_path}")
        return None

    # Load and preprocess the image
    transform = get_transforms('test')

    try:
        image = Image.open(image_path).convert('RGB')
        input_tensor = transform(image).unsqueeze(0).to(device)

        # Set model to evaluation mode
        model.eval()

        # Make prediction
        with torch.no_grad():
            output = model(input_tensor)
            probabilities = torch.nn.functional.softmax(output[0], dim=0)
            confidence, prediction = torch.max(probabilities, 0)

        # Prepare result
        result = {
            'class': classes[prediction.item()],
            'confidence': confidence.item(),
            'is_damaged': prediction.item() != 1,  # Not normal means damaged
            'is_road': True  # Our model assumes all inputs are roads
        }

        # Display the image with the prediction
        plt.figure(figsize=(10, 6))
        plt.imshow(image)

        if result['is_damaged']:
            damage_status = f"DAMAGED ROAD: {result['class']}"
            title_color = 'red'
        else:
            damage_status = "NORMAL ROAD: No Damage Detected"
            title_color = 'green'

        plt.title(f"{damage_status}\nConfidence: {result['confidence']:.2f}",
                fontsize=14, color=title_color)
        plt.axis('off')
        plt.show()

        return result
    except Exception as e:
        print(f"Error during prediction: {e}")
        return None

# Function to validate dataset structure and find the correct path
def find_dataset_path(base_path):
    """
    Search for the dataset structure in the given path and return the correct path

    Args:
        base_path: The base directory to search in

    Returns:
        The correct dataset path, or None if not found
    """
    print(f"Searching for dataset in {base_path}...")

    # Check if base path exists
    if not os.path.exists(base_path):
        print(f"Error: Base path {base_path} does not exist.")
        return None

    # Define the expected classes
    classes = ['crack', 'normal', 'pothole']

    # Check if any of the expected classes exist directly in the base path
    direct_class_count = 0
    for class_name in classes:
        if os.path.exists(os.path.join(base_path, class_name)):
            direct_class_count += 1

    if direct_class_count >= 2:  # At least 2 classes found directly in base_path
        print(f"Found dataset with {direct_class_count} classes at {base_path}")
        return base_path

    # List all subdirectories in the base path
    subdirs = [d for d in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, d))]

    # Check each subdirectory for dataset structure
    for subdir in subdirs:
        subdir_path = os.path.join(base_path, subdir)
        class_count = 0
        for class_name in classes:
            if os.path.exists(os.path.join(subdir_path, class_name)):
                class_count += 1

        if class_count >= 2:  # At least 2 classes found in this subdirectory
            print(f"Found dataset with {class_count} classes at {subdir_path}")
            return subdir_path

    # Search up to 2 levels deep
    for subdir in subdirs:
        subdir_path = os.path.join(base_path, subdir)
        subsubdirs = [d for d in os.listdir(subdir_path) if os.path.isdir(os.path.join(subdir_path, d))]

        for subsubdir in subsubdirs:
            subsubdir_path = os.path.join(subdir_path, subsubdir)
            class_count = 0
            for class_name in classes:
                if os.path.exists(os.path.join(subsubdir_path, class_name)):
                    class_count += 1

            if class_count >= 2:  # At least 2 classes found in this subdirectory
                print(f"Found dataset with {class_count} classes at {subsubdir_path}")
                return subsubdir_path

    print("Could not find a proper dataset structure in the given path.")
    print("Please make sure your dataset has directories named 'crack', 'normal', and 'pothole'.")
    print("Or specify the correct path to the dataset.")

    # Return the base path anyway, the dataset class will handle missing data
    return base_path

# Main execution function
def main():
    # Mount Google Drive if needed
    mount_drive()

    # Set initial base path
    base_dir = '/content'

    # Check if dataset is in Drive
    drive_path = '/content/drive/MyDrive'
    if os.path.exists(drive_path):
        # Look for dataset in Drive
        for item in os.listdir(drive_path):
            if 'road' in item.lower() and 'damage' in item.lower():
                zip_path = os.path.join(drive_path, item)
                if zip_path.endswith('.zip'):
                    print(f"Found potential dataset ZIP: {zip_path}")
                    extract_to = '/content/road_damage_dataset'
                    extract_dataset(zip_path, extract_to)
                    base_dir = extract_to
                    break
                elif os.path.isdir(os.path.join(drive_path, item)):
                    print(f"Found potential dataset directory: {os.path.join(drive_path, item)}")
                    base_dir = os.path.join(drive_path, item)
                    break

    # Try to find the dataset structure
    data_dir = find_dataset_path(base_dir)

    if not data_dir:
        print("Dataset not found automatically. Please enter the path to your dataset:")
        user_path = input("Dataset path: ").strip()
        if user_path:
            data_dir = user_path
        else:
            data_dir = base_dir  # Default to base_dir if no input

    print(f"\nUsing dataset path: {data_dir}")

    # Define transformations
    data_transforms = {
        'train': get_transforms('train'),
        'val': get_transforms('val'),
        'test': get_transforms('test')
    }

    # Create datasets with error handling
    try:
        image_datasets = {
            'train': RoadDamageDataset(root_dir=data_dir, transform=data_transforms['train'], subset='train'),
            'val': RoadDamageDataset(root_dir=data_dir, transform=data_transforms['val'], subset='val'),
            'test': RoadDamageDataset(root_dir=data_dir, transform=data_transforms['test'], subset='test')
        }

        # Verify datasets have data
        empty_datasets = []
        for subset, dataset in image_datasets.items():
            if len(dataset) == 0:
                empty_datasets.append(subset)

        if empty_datasets:
            print(f"Warning: The following datasets are empty: {', '.join(empty_datasets)}")

            # If training set is empty, we cannot proceed
            if 'train' in empty_datasets:
                print("Training dataset is empty. Cannot train the model.")
                return
    except Exception as e:
        print(f"Error creating datasets: {e}")
        return

    # Create dataloaders with appropriate batch size and error handling
    batch_size = 32
    try:
        dataloaders = {
            'train': DataLoader(image_datasets['train'],
                               batch_size=min(batch_size, len(image_datasets['train'])),
                               shuffle=True,
                               num_workers=2),
            'val': DataLoader(image_datasets['val'],
                             batch_size=min(batch_size, max(1, len(image_datasets['val']))),
                             shuffle=False,
                             num_workers=2),
            'test': DataLoader(image_datasets['test'],
                              batch_size=min(batch_size, max(1, len(image_datasets['test']))),
                              shuffle=False,
                              num_workers=2)
        }
    except Exception as e:
        print(f"Error creating data loaders: {e}")
        return

    # Get class names
    class_names = image_datasets['train'].classes

    # Set device
    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    # Initialize model
    model = RoadDamageViT(num_classes=len(class_names))

    # Define loss function and optimizer
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(model.parameters(), lr=0.0001, weight_decay=0.01)

    # Train model
    try:
        trained_model, history = train_model(
            model,
            dataloaders={k: dataloaders[k] for k in ['train', 'val']},
            criterion=criterion,
            optimizer=optimizer,
            num_epochs=20,
            device=device
        )

        # Plot training history
        plot_training_history(history)
    except Exception as e:
        print(f"Error during training: {e}")
        return

    # Evaluate on test set if available
    if len(image_datasets['test']) > 0:
        try:
            test_acc, _ = evaluate_model(
                trained_model,
                dataloaders['test'],
                criterion,
                device=device,
                classes=class_names
            )
        except Exception as e:
            print(f"Error during evaluation: {e}")
    else:
        print("Test dataset is empty. Skipping evaluation.")

    # Save the model
    try:
        torch.save(trained_model.state_dict(), '/content/drive/MyDrive/road_damage_vit_model.pth')
        print("Model saved as 'road_damage_vit_model.pth'")
    except Exception as e:
        print(f"Error saving model: {e}")

    # Try to find a test image for demonstration
    test_image_path = None
    # Look in test folder
    if len(image_datasets['test'].samples) > 0:
        test_image_path = image_datasets['test'].samples[0][0]
    # Or look anywhere else
    else:
        for root, _, files in os.walk(data_dir):
            for file in files:
                if file.endswith(('.jpg', '.jpeg', '.png')):
                    test_image_path = os.path.join(root, file)
                    break
            if test_image_path:
                break

    # Example usage for single image prediction
    if test_image_path:
        try:
            print(f"\nTesting prediction on image: {test_image_path}")
            result = predict_image(trained_model, test_image_path, device)
            if result:
                print("\nPrediction Result:")
                print(f"Class: {result['class']}")
                print(f"Confidence: {result['confidence']:.2f}")
                print(f"Road Status: {'Damaged' if result['is_damaged'] else 'Normal'}")
        except Exception as e:
            print(f"Error during prediction: {e}")
    else:
        print("\nNo test image found. Please provide an image path to test the model.")
        print("Example usage:")
        print("result = predict_image(trained_model, '/path/to/your/image.jpg', device='cuda')")

if __name__ == "__main__":
    try:
        main()
    except Exception as e:
        print(f"An error occurred in the main function: {e}")
        import traceback
        traceback.print_exc()
```

```python
import torch
import torch.nn as nn
from torchvision.models import vit_b_16, ViT_B_16_Weights
from PIL import Image
import matplotlib.pyplot as plt
import torchvision.transforms as transforms

# Define the model class
class RoadDamageViT(nn.Module):
    def __init__(self, num_classes=3):
        super(RoadDamageViT, self).__init__()
        self.model = vit_b_16(weights=ViT_B_16_Weights.IMAGENET1K_V1)
        in_features = self.model.heads.head.in_features
        self.model.heads.head = nn.Linear(in_features, num_classes)

    def forward(self, x):
        return self.model(x)

# Load model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = RoadDamageViT()
model.load_state_dict(torch.load('/content/drive/MyDrive/road_damage_vit_model.pth', map_location=device))
model.to(device)
model.eval()

# Transform
def get_transforms():
    return transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406],
                             std=[0.229, 0.224, 0.225])
    ])

# Predict function
def predict_image(image_path):
    classes = ['Crack', 'Normal (No Damage)', 'Pothole']
    image = Image.open(image_path).convert('RGB')
    input_tensor = get_transforms()(image).unsqueeze(0).to(device)

    with torch.no_grad():
        output = model(input_tensor)
        probabilities = torch.nn.functional.softmax(output[0], dim=0)
        confidence, prediction = torch.max(probabilities, 0)

    plt.imshow(image)
    plt.title(f"{classes[prediction]} ({confidence:.2f})")
    plt.axis('off')
    plt.show()

    return classes[prediction.item()], confidence.item()

# Example usage:
# predict_image('/path/to/your/image.jpg')
```
